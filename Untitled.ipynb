* Step 3: Rename Columns for Consistency and renaming confusing columns

```python
df.rename(columns={
    'Hipertension': 'Hypertension',
    'Handcap': 'Handicap',
    'No-show': 'NoShow'
}, inplace=True)

# Show new column names
df.columns
```

```python
df.head() 
```

* Step 4: Convert ScheduledDay and AppointmentDay to datetime

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

# Check updated data types
df[['ScheduledDay', 'AppointmentDay']].dtypes
```

* What does datetime64[ns] mean?

_datetime64 → This means the values are now real datetime objects, not just strings.

[ns] → The resolution is nanoseconds (standard for pandas).

This allows you to do math with dates, extract weekdays/months, etc.

_UTC stands for Coordinated Universal Time (a global time standard).

Your dates are timezone-aware: they include UTC time zone info

```python
# Drop the WaitDays column
df.drop('WaitDays', axis=1, inplace=True)
```

* Step 5: Check for Missing Values

```python
df.isnull().sum()
```

this output says that there is no missing values

* Step 6: Check and remove duplicate rows if any

```python
print("Duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

* Step 7: Remove rows with invalid ages (e.g., negative age)


```python
df = df[df['Age'] >= 0]

# Check age stats
df['Age'].describe()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace('-', '_')

df.head() 
```

* Step 8: Drop columns not useful for analysis

```python
df.drop(['PATIENTID', 'APPOINTMENTID'], axis=1, inplace=True)

# View remaining columns
df.columns
```

* Step 9: Check Class Distribution (Target Variable)

```python
#Check how many patients showed up vs didn't

df['NOSHOW'].value_counts()
```
